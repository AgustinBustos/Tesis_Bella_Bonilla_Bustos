```python
def binaryGrapher(x,y,df,line='no'):  
  #print('--------------------------------------------df--------------------------')
  #print(df,'este es df')
  a=y[0]
  b=y[1]
  #print('a es',a)
  #print('b es',b)
  aux1=df[df.columns.values]
  aux2=df[df.columns.values]
  
  df0=aux1
  df0['color']=a
  df0['why']=aux1.loc[:,a]

  df1=aux2

  df1['color']=y[1]
  df1['why']=aux2.loc[:,y[1]]
  #print('--------------------------------------------auxiliares--------------------------')

  #print(df0,df1,'estos son los dfs')
  result = df0.append(df1)
  #print('--------------------------------------------resultado--------------------------')

  #print(result)
  if line=='yes':
    px.line(result,x=x,y='why',color='color').show()
  else:
    px.scatter(result,x=x,y='why',color='color').show()
def rellenator1(variable,weights,countryDf,k,graph): #weights es distance o uniform
  #necesito una funcion que te diga si hay muhos nans (mas del 50 porciento)???????????????????????????????????????
  
    PreTrain=countryDf[['Year',variable]].dropna()

    X1=PreTrain[['Year']].to_numpy()
    y1=PreTrain[variable].to_numpy()
    Xpred=countryDf[['Year']].to_numpy()
    neigh = KNeighborsRegressor(n_neighbors=k,weights=weights)
    neigh.fit(X1, y1)
    yhat=neigh.predict(Xpred)
    Relleno=countryDf[countryDf.columns.values]
    Relleno[f'new{variable}']=yhat
    #creo que funciono al fin
    if graph:
      print(np.count_nonzero(np.isnan(y1)))
      binaryGrapher('Year',[variable,f'new{variable}'],Relleno)
      
    return Relleno

def fullRellenator(rellenator,variable,df,boroughArray,graph):

  data=df.copy()
  for i in boroughArray:
    if graph:
      print(i)
    
    data.loc[data['Borough']==i,variable]=rellenator(variable,'distance',data.loc[data['Borough']==i],3,graph).loc[:,f'new{variable}']

  return data
```
