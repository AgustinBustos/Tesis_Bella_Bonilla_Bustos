```python
def normByYear(xt):
  dummies=xt[:,0:32]
  pOutputs=xt[:,32:42]
  pMean=torch.mean(pOutputs,dim=-1)
  pStd=torch.std(pOutputs,dim=-1)
  #torch.outer(dummies,pOutputs)
  unos=torch.ones(10)
  unos=to_device(unos,device)
  #pOutputs-pMean
  kronekerMean=torch.outer(pMean,unos)
  kronekerStd=torch.outer(pStd,unos)
  newpOutputs=(pOutputs-kronekerMean)/kronekerStd
  dummies=xt[0,0:32]
  outNormed=newpOutputs[0]
  conecction=torch.outer(dummies,outNormed)
  conecction=conecction.reshape(-1)
  newFirstRow=torch.cat((dummies,outNormed,conecction),dim=-1)
  acum=newFirstRow.reshape(1,-1)
  for i in range(len(xt)-1):
    dummies=xt[i+1,0:32]
    outNormed=newpOutputs[i+1]
    conecction=torch.outer(dummies,outNormed)
    conecction=conecction.reshape(-1)
    newFirstRow=torch.cat((dummies,outNormed,conecction),dim=-1)
    acum=torch.cat((acum,newFirstRow.reshape(1,-1)),dim=0)
  return acum,pMean.reshape(-1,1),pStd.reshape(-1,1)
```

```python

#entreno el modelo
# Utility function to train the model
def fit(epochNum, model, error, opt, xt,yt,xv,yv,graph=True,valCheck=True,scheduler=False,regMode=True,isMetric=False):
    if isMetric==False:
      metric=error
    else:
      metric=isMetric

    if regMode:
      model.eval()
    modelCopy=copy.deepcopy(model)
    vFuture=metric(model(xv), yv).item()
    vPast=vFuture+0.1
    tFuture=error(model(xt), yt).item()
    tPast=tFuture+0.1
    tLoss=[tFuture]
    vLoss=[vFuture]
    # Repeat for given number of epochs
    oldProg=0
    vWeights=model.state_dict()
    vUpToNowMin=vPast
    tWeights=model.state_dict()
    tUpToNowMin=tPast
    
    #loop
    for epoch in range(epochNum):
      if regMode:
        model.train()
      yhat = model(xt)
      loss = error(yhat, yt)
      opt.zero_grad()
      loss.backward()
      opt.step()
      if scheduler:    
        scheduler.step()
        
        #actualizamos
      if regMode:
        model.eval()
        tPast=tFuture
        tFuture=error(model(xt), yt).item()
      else:
        tPast=tFuture
        tFuture=loss.item()
      if valCheck:
        vPast=vFuture
        vFuture=metric(model(xv), yv).item()
        
        #guardamos parametros del modelo si es menor
      if valCheck:
        if vFuture<vUpToNowMin:
          vWeights=copy.deepcopy(model.state_dict())
          vUpToNowMin=vFuture
            
            

            

        
      if tFuture<tUpToNowMin:
        tWeights=copy.deepcopy(model.state_dict())
        tUpToNowMin=tFuture
        #metemos las losses en una lista
      tLoss.append(tFuture)
      if valCheck:
        vLoss.append(vFuture)
      if graph==False:
          pass
      else:
        progress=epoch/epochNum #puedo crear un estimated time
        newProg=int(progress*len(anime))
        if oldProg!=newProg:
          print(anime[newProg], end='')
          oldProg=newProg

      
      
      
      
    tlossDf=pd.DataFrame({'error':tLoss,'time':np.arange(0,epochNum+1,1)})
    tlossDf['type']='train'
    if valCheck:
      vlossDf=pd.DataFrame({'error':vLoss,'time':np.arange(0,epochNum+1,1)})
      vlossDf['type']='validation'
      dat=pd.concat([tlossDf,vlossDf])
      
    if graph:
      if valCheck:
        px.line(dat,x='time',y='error',color='type').show()
          
          
      else:
        px.line(tlossDf,x='time',y='error').show()

    
    if regMode:
      model.eval()
    return {'trainError': error(model(xt), yt).item(),
            'valError': metric(model(xv), yv).item(),
            'minVal':np.amin(np.array(vLoss)),
            'minTrain':np.amin(np.array(tLoss)),
            'vWeights':vWeights,
            'tWeights':tWeights,
            'lastWeights':model.state_dict(),
            'tLoss':tLoss,
            'vLoss':vLoss}
```

```python
#new fit, lr selector, las mini epoch son 100
def lrLinspace(xt,yt,xv,yv,error,modelToCopy,optFunction,valError=False,graph=True,miniEpoch=100,minLr=1e-7,negative=False,upperBound=0.1,lossDiffLowerBound=1e-10,smoothNeg=False,move=20): #?????????????????hacer la val version
  #calculo el linspace de lr
  model=copy.deepcopy(modelToCopy)
  
  opt=optFunction(model,minLr)


  a=np.exp((np.log(1/minLr))/miniEpoch)
  lrsFuction=lambda epoch: a**(epoch-move)   #le cambio el numero restando para mover el linspace
  if negative:
    lrsFuction=lambda epoch: -(a**(epoch-move))
  

  scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda=lrsFuction)
  tFuture=error(model(xt), yt).item()
  tPast=tFuture+0.1
  tLoss=[tFuture]
  if valError:
    vFuture=error(model(xv), yv).item()
    vPast=vFuture+0.1
    vLoss=[vFuture]

  
  for epoch in range(miniEpoch):
    yhat = model(xt)
    loss = error(yhat, yt)
    opt.zero_grad()
    loss.backward()
    opt.step()

    scheduler.step()

    tLoss.append(loss.item())
    if valError:
      vLoss.append(error(model(xv),yv).item())

  #creo las funciones de diferenciacion
  lrs=[lrsFuction(x)*minLr for x in np.arange(0,miniEpoch+1,1)]
  if negative:
    lrs=[-(lrsFuction(x)*minLr) for x in np.arange(0,miniEpoch+1,1)]

  lossDiff=[]
  for i in range(miniEpoch):
    lossDiff.append(tLoss[i+1]-tLoss[i])

  if graph:
    
    if valError:
      tlossDf=pd.DataFrame({'error':tLoss,'lrs':lrs})
      tlossDf['type']='train'
      vlossDf=pd.DataFrame({'error':vLoss,'lrs':lrs})
      vlossDf['type']='validation'
      dat=pd.concat([tlossDf,vlossDf])
      px.line(dat,x='lrs',y='error',color='type',log_x=True).show()
      
    else:
      px.line(x=lrs,y=tLoss,log_x=True).show()
      px.line(x=lrs[1:len(lrs)],y=lossDiff,log_x=True).show()

    

  #obtengo minimos de manera analitica
  #1] mal
  min10=lrs[tLoss.index(min(tLoss))]/10
  #2]mejor
  future=lossDiff[0]
  past=future+0.1
  counter=1
  while future<past:
    past=future
    future=lossDiff[counter]
    counter=counter+1
    
  minSteep=lrs[counter-2] #le resto 2 por el tema de las derivadas
  #3]antes de que explote
  beforeExplotionLr=lrs[beforeExplotion(tLoss)]
  #4]el tema del negative es delicado
  if negative:
    future=tLoss[0]
    counter=1
    while future<tLoss[0]+upperBound:
      future=tLoss[counter]
      counter=counter+1
      
    negLr=lrs[counter-1] 
    if smoothNeg:
      negLr=negLr/10
  else:
    negLr=lrs[0]  #juego seguro, hasta ahora esta bien
  if negative:
    model1=copy.deepcopy(modelToCopy)
    
    opt1=optFunction(model1,1)
    lrLmbda=lambda epoch: -negLr #cycleFunct(epoch,100,negLr,2,2)
    scheduler1 = torch.optim.lr_scheduler.LambdaLR(opt1, lr_lambda=lrLmbda)
    tLoss2=[]
    for epoch in range(miniEpoch):
      yhat1 = model1(xt)
      loss1 = error(yhat1, yt)
      opt1.zero_grad()
      loss1.backward()
      opt1.step()
      scheduler1.step()
      tLoss2.append(loss1.item())
    if graph:
      px.line(x=np.arange(0,miniEpoch,1),y=tLoss2).show()
    #obtengo la q de epochs antes de la explosion
    future1=tLoss2[0]
    counter1=1
    while future1<tLoss2[0]+upperBound:
      future1=tLoss2[counter1]
      counter1=counter1+1
      
    maxEpochDuration=counter1
    if graph:
      print('negLr: ',negLr,'upperBound: ',upperBound, 'maxEpochDuration: ',maxEpochDuration)
  else:
    maxEpochDuration=0
  #5]busco la minima lr que tenga sentido para aplicar un ultra deep training en la epoch final
  indicesOfMin=[]
  try:
    for i in range(beforeExplotion(tLoss)):
      if lossDiff[i]>lossDiff[i+1]:
        if lossDiff[i+1]<lossDiff[i+2]:
          indicesOfMin.append(i)
    
    #una vez que tengo los minimos locales, chequeo cual pega el mayor salto
    
    indicesOfMinDiff=np.diff(np.array(indicesOfMin))   
    bigJump=np.argmax(indicesOfMinDiff)+1    #sumarle 1 es extremadamente importante
    delicateIndex=indicesOfMin[bigJump]
  except:
    delicateIndex=0

  
  delicateLr=lrs[delicateIndex]
  #6] la primera vez que se hace positiva la diff
  firstPositiveCounter=0
  while lossDiff[firstPositiveCounter]<0:
    firstPositiveCounter=firstPositiveCounter+1
  firstPositive=lrs[firstPositiveCounter]
  #7] hago el steep per para val
  if valError:
    vLossDiff=np.diff(np.array(vLoss))
    future=vLossDiff[0]
    past=future+0.1
    counter=1
    while future<past:
      past=future
      future=vLossDiff[counter]
      counter=counter+1
      
    vMinSteep=lrs[counter-2] #le resto 2 por el tema de las derivadas
  else:
    vMinSteep=minLr
  #8]hago el before explotion para val
  if valError:
    vBeforeExplotionLr=lrs[beforeExplotion(vLoss)]
  else:
    vBeforeExplotionLr=minLr

  resultados={'min/10': min10,
          'minSteep':minSteep,
          'beforeExplotionLr':beforeExplotionLr,
          'negative/duration':[negLr,maxEpochDuration],
          'firstPositive':firstPositive,
          'delicateLr':delicateLr,
          'vMinSteep':vMinSteep,
          'vBeforeExplotionLr':vBeforeExplotionLr}
          #'tLoss':tLoss}
  if graph:
    print('Los resultados son: ',resultados)
    print('////////////////////////////////////////////////////////////////////////////')

  
  return resultados


```

```python
#la opt function debe tener 2 argumentos, model, lr, y tienen que estar en ese orden, y tienen que ser parametrizados con def f(model=None,minLr=None)
def cyclicFit(semiEpochs,processIteration, model, error, optFunction, xt,yt,xv,yv,cyclerFunction=cycleFunct,fromScratch=True,fullGraphList=False,graph=True,early=False,valCheck=True,fullGraph=False,anyGraph=True,n=0):
  if fullGraph:
    fullGraphList=[True for i in range(processIteration+1)]

  if fullGraphList==False:
    fullGraphList=[False for i in range(processIteration+1)]
  result=[]
  vfirstError=[error(model(xv),yv).item()]
  tfirstError=[error(model(xt),yt).item()]
  if anyGraph==False:
    graph=False
  def maxLrCut(t,j):
    if t<=j:
      return 'minSteep'
    else:
      return 'beforeExplotionLr'
  lastError=[error(model(xt),yt).item()]

  #empiezo el entrenamiento  
  if fromScratch:
    maxLr=lrLinspace(xt,yt,xv,yv,error,model,optFunction,fullGraphList[0])[maxLrCut(0,n)] #notar que graph es una lista de trues y falses
    negLr=lrLinspace(xt,yt,xv,yv,error,model,optFunction,fullGraphList[0],negative=True,upperBound=lastError[0]+0.1)['negative/duration']
    
    #una vez que encontre el optimizador, le paso un ciclo
    lrsFuction=lambda epoch: cyclerFunction(epoch,semiEpochs,maxLr,0,negLr)
    opt=optFunction(model,1) #notar esta lr de 1 para que funcione el scheduler
    scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda=lrsFuction)
    if fullGraphList[0]:
      px.line(x=np.arange(0,semiEpochs,1),y=[cyclerFunction(x,semiEpochs,maxLr,0,negLr) for x in np.arange(0,semiEpochs,1) ]).show()
    if graph:
      fullGraphList[0]=True
    if fullGraphList[0]:
      print('--------------------------------------------------')
      print('El maximo Lr a usar es: ',maxLr)
      print('upperBound: ',lastError[0]+0.1)
      print('--------------------------------------------------')
    result.append(fit(semiEpochs, model, error, opt, xt,yt,xv,yv,graph=fullGraphList[0],scheduler=scheduler))
    lastError.append(error(model(xt),yt).item())
  
  #ahora hago el loop de iteraciones  
  if early==False:
    
    for i in range(processIteration):
      
      maxLr=lrLinspace(xt,yt,xv,yv,error,model,optFunction,fullGraphList[i+1])[maxLrCut(i+1,n)] #notar que graph es una lista de trues y falses
      negLr=lrLinspace(xt,yt,xv,yv,error,model,optFunction,fullGraphList[0],negative=True,upperBound=lastError[i]/2)['negative/duration']
      #una vez que encontre el optimizador, le paso un ciclo
      lrsFuction=lambda epoch: cyclerFunction(epoch,semiEpochs,maxLr,i+1,negLr)
      opt=optFunction(model,1) #notar esta lr de 1 para que funcione el scheduler
      scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda=lrsFuction)
      if fullGraphList[i+1]:
        px.line(x=np.arange(0,semiEpochs,1),y=[cyclerFunction(x,semiEpochs,maxLr,i+1,negLr) for x in np.arange(0,semiEpochs,1) ]).show()
      if graph:
        fullGraphList[i+1]=True
      if fullGraphList[i+1]:
        print('--------------------------------------------------')
        print('El maximo Lr a usar es: ',maxLr)
        print(maxLrCut(i+1,n))
        print('upperBound: ',lastError[i])
        print('--------------------------------------------------')
      result.append(fit(semiEpochs, model, error, opt, xt,yt,xv,yv,graph=fullGraphList[i+1],scheduler=scheduler))
      lastError.append(error(model(xt),yt).item())
  elif early==True:  #hago el mismo loop pero usando early stopping
    vFuture=error(model(xv),yv).item()
    vPast=vFuture+0.1
    counter=1

    while vFuture<vPast:
      maxLr=lrLinspace(xt,yt,xv,yv,error,model,optFunction,fullGraph)[maxLrCut(i+1,n)] #notar que graph es una lista de trues y falses
      #una vez que encontre el optimizador, le paso un ciclo
      lrsFuction=lambda epoch: cyclerFunction(epoch,semiEpochs,maxLr,counter)
      opt=optFunction(model,1) #notar esta lr de 1 para que funcione el scheduler
      scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda=lrsFuction)
      if fullGraph:
        px.line(x=np.arange(0,semiEpochs,1),y=[cyclerFunction(x,semiEpochs,maxLr,counter) for x in np.arange(0,semiEpochs,1) ]).show()
      if graph:
        print('--------------------------------------------------')
        print('El maximo Lr a usar es: ',maxLr)
        print(maxLrCut(i+1,n))
        print('--------------------------------------------------')
      result.append(fit(semiEpochs, model, error, opt, xt,yt,xv,yv,graph=graph,scheduler=scheduler))

      #actualizo valores del early sttoping
      vPast=vFuture
      vFuture=error(model(xv),yv).item() 


  #hago un grafico de cada semiepoch    
  tLosses=tfirstError+[x['trainError'] for x in result]
  vLosses=vfirstError+[x['valError'] for x in result]
  tlossDf=pd.DataFrame({'error':tLosses,'time':np.arange(0,len(tLosses),1)})
  tlossDf['type']='train'
  if valCheck:
    vlossDf=pd.DataFrame({'error':vLosses,'time':np.arange(0,len(vLosses),1)})
    vlossDf['type']='validation'
    dat=pd.concat([tlossDf,vlossDf])     
  if anyGraph:
    if valCheck:
      px.scatter(dat,x='time',y='error',color='type').show()       
    else:
      px.scatter(tlossDf,x='time',y='error').show()



  return {'tLosses':[x['trainError'] for x in result], 'vLosses':[x['valError'] for x in result]}





def compoundCycle(x):# tengo que hacerlo
  return x
```

```python
#la opt function debe tener 2 argumentos, model, lr, y tienen que estar en ese orden, y tienen que ser parametrizados con def f(model=None,minLr=None)
def tunedCyclicFit(parameterSelectorArray,cyclerFunctionArray,semiEpochsArray, model, error, optFunction, xt,yt,xv,yv,graph=True,early=False,valCheck=True,fullGraph=False,semiEarly=False):
  result=[]
  tFuture=error(model(xt),yt).item()
  vFuture=error(model(xv),yv).item()
  tPast=tFuture+0.1
  vPast=vFuture+0.1
  vError=[vFuture]
  tError=[tFuture]
  


  #empiezo el entrenamiento  
  for i in range(len(semiEpochsArray)): 
    #instancio las funciones
    lrLinspace=parameterSelectorArray[i]
    cyclerFunction=cyclerFunctionArray[i]
    semiEpochs=semiEpochsArray[i]
    if graph:
      printmd(f'<h1>Cycle Change {i}<h1/>')
      printmd('<h3>BUSCO LOS PARAMETROS DEL CYCLE<h3/>')
    

    

    #selecciono los parametros del ciclo
    cycleParams=lrLinspace(xt,yt,xv,yv,error,model,optFunction) 
    #una vez que encontre los parametros, le paso un ciclo
    if graph:
      printmd('<h3>CREO LA FUNCION DE CYCLE<h3/>')
    lrsFuction=lambda epoch: cyclerFunction(epoch,semiEpochs,cycleParams)
    opt=optFunction(model,1) #notar esta lr de 1 para que funcione el scheduler
    scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda=lrsFuction)
    if fullGraph:
      px.line(x=np.arange(0,semiEpochs,1),y=[cyclerFunction(x,semiEpochs,cycleParams) for x in np.arange(0,semiEpochs,1) ]).show()
    if graph:
      print('--------------------------------------------------')
      print('Los parametros a usar son: ',cycleParams)
      print('--------------------------------------------------')

    if graph:
      printmd('<h3>FITTEO<h3/>')
    if semiEarly==False:
      result.append(fit(semiEpochs, model, error, opt, xt,yt,xv,yv,graph=graph,scheduler=scheduler))
    elif semiEarly=='minTrain':
      resultado=fit(semiEpochs, model, error, opt, xt,yt,xv,yv,graph=graph,scheduler=scheduler)
      result.append(resultado)
      model.load_state_dict(resultado['tWeights'])

      #la pregunta es si tiro el mismo fit, va a cambiar el modelo?
    elif semiEarly=='minVal':
      resultado=fit(semiEpochs, model, error, opt, xt,yt,xv,yv,graph=graph,scheduler=scheduler)
      result.append(resultado)
      model.load_state_dict(resultado['vWeights'])
      
     



    #llevo la cuenta
    tPast=tFuture
    vPast=vFuture
    tFuture=error(model(xt),yt).item()
    vFuture=error(model(xv),yv).item()
    vError.append(vFuture)
    tError.append(tFuture)
    if graph:
      printmd('<hr/>')


    if early:
      if vPast<vFuture:
        break
    

  #hago un grafico de cada semiepoch    
  tLosses=tError
  vLosses=vError
  tlossDf=pd.DataFrame({'error':tLosses,'time':np.arange(0,len(tLosses),1)})
  tlossDf['type']='train'
  if valCheck:
    vlossDf=pd.DataFrame({'error':vLosses,'time':np.arange(0,len(vLosses),1)})
    vlossDf['type']='validation'
    dat=pd.concat([tlossDf,vlossDf])     
  if graph:
    if valCheck:
      px.scatter(dat,x='time',y='error',color='type').show()       
    else:
      px.scatter(tlossDf,x='time',y='error').show()
  

  #final
  finalTLosses=[x['trainError'] for x in result]
  finalVLosses=[x['valError'] for x in result]
  finalMinTrain=min([x['minTrain'] for x in result])
  finalMinVal=min([x['minVal'] for x in result])
  #mejor modelo
  minTrainIndex=[x['minTrain'] for x in result].index(finalMinTrain)
  minValIndex=[x['minVal'] for x in result].index(finalMinVal)
  minTrainResult=result[minTrainIndex]
  minValResult=result[minValIndex]
  minTrainModel=minTrainResult['tWeights']
  minValModel=minValResult['vWeights']


  
  return {'tLosses':finalTLosses, 
          'vLosses':finalVLosses,
          'minTrain':finalMinTrain,
          'minVal':finalMinVal,
          'tWeights':minTrainModel,
          'vWeights':minValModel,
          'results':result}

          





```

```python
def wdLinspace(model,optFunction,error, xt,yt,xv,yv,maxLr=0.003,minWd=100,repetitions=1000,num=100,graph=True):
  wdSpace=np.logspace(1,minWd,num=num,base=0.5)
  newTError=[]
  newVError=[]

  for wd in wdSpace:
    
    modelCopy=copy.deepcopy(model)
    opt=optFunction(modelCopy,wd)
    errors=simpleCycle(repetitions,maxLr,modelCopy,opt,error,xt,yt,xv,yv,graph=False)
    #le voy a pasar un ciclo
    
    newTError.append(errors['trainError'])
    newVError.append(errors['valError'])

  trainDf=pd.DataFrame({'error':newTError,'wd':wdSpace})
  trainDf['type']='train'
  valDf=pd.DataFrame({'error':newVError,'wd':wdSpace})
  valDf['type']='val'

  dat=pd.concat([trainDf,valDf])
  if graph:
    px.line(dat,x='wd',y='error',color='type',log_x=True).show()

  indexOfMin=np.argmin(np.array(newVError))
  return wd[indexOfMin]
  



```

```python
def simpleCycle(semiEpochs,maxLr,model,opt,error,xt,yt,xv,yv,graph=True,isMetric=False):

  lrsFuction=lambda epoch: cycleFunct(epoch,semiEpochs,maxLr,5,4)
  scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_lambda=lrsFuction)
  if graph:
    px.line(x=np.arange(0,semiEpochs,1),y=[cycleFunct(x,semiEpochs,maxLr,5,4) for x in np.arange(0,semiEpochs,1) ]).show()
  result=fit(semiEpochs, model, error, opt, xt,yt,xv,yv,graph=graph,scheduler=scheduler,isMetric=isMetric)
  return result
```

```python
#cycle stochastic
#la idea de hacer resampling es que le paso un numero, y en esa semiepoch se queda esperando hasta que los otrso modelos corran, luego
# verifica quien tiene menos train error o val error, y apuesta a ese, y luego sigue
#quiero que me devuelvan los graficos del training, que japa
def stochCycle(stochs,parameterSelectorArray,cyclerFunctionArray,semiEpochsArray, modelFunction, error, optFunction, xt,yt,xv,yv,whichError='minTrain',early=False,valCheck=True,semiEarly=False):
  if whichError=='minTrain':
    weights='tWeights'
  else:
    weights='vWeights'
  resultsArray=[]
  for i in range(stochs):
    model=modelFunction()
    to_device(model, device)
    resultado=tunedCyclicFit(parameterSelectorArray,cyclerFunctionArray,semiEpochsArray, model, error, optFunction, xt,yt,xv,yv,graph=False,early=early,valCheck=valCheck,fullGraph=False,semiEarly=semiEarly)
    resultsArray.append(resultado)

  errores=[x[whichError] for x in resultsArray]
  minimoError=min(errores)
  indexMin=errores.index(minimoError)
  bestResult=resultsArray[indexMin]
  
  return {'array':errores,'minError':minimoError,'bestModel':bestResult[weights]}
def cycle1Tree(weightsSelectorFunct,whentoCut,howManyStochs,parameterSelectorArray,cyclerFunctionArray,semiEpochsArray, modelFunction, error, optFunction, xt,yt,xv,yv,whichError='minTrain',graph=True,early=False,valCheck=True,fullGraph=False,semiEarly=False):
  #corto las array en 2
  totalSemi=len(parameterSelectorArray)
  parameterSelectorArray1=parameterSelectorArray[0:whentoCut]
  parameterSelectorArray2=parameterSelectorArray[whentoCut:totalSemi]
  cyclerFunctionArray1=cyclerFunctionArray[0:whentoCut]
  cyclerFunctionArray2=cyclerFunctionArray[whentoCut:totalSemi]
  semiEpochsArray1=semiEpochsArray[0:whentoCut]
  semiEpochsArray2=semiEpochsArray[whentoCut:totalSemi]

  #obtengo el mejor modelo

  weights=weightsSelectorFunct(howManyStochs,parameterSelectorArray1,cyclerFunctionArray1,semiEpochsArray1, modelFunction, error, optFunction, xt,yt,xv,yv,whichError=whichError,early=early,valCheck=valCheck,semiEarly=semiEarly)
  model=modelFunction()
  to_device(model, device)
  model.load_state_dict(weights['bestModel'])
  resultadoFinal=tunedCyclicFit(parameterSelectorArray2,cyclerFunctionArray2,semiEpochsArray2, model, error, optFunction, xt,yt,xv,yv,graph=True,early=False,valCheck=True,fullGraph=False,semiEarly=False)
  return resultadoFinal

  
def negativeCaruanaResampling(stochs,parameterSelectorArray,cyclerFunctionArray,semiEpochsArray, modelFunction, error, optFunction, xt,yt,xv,yv,whichError='minTrain',early=False,valCheck=True,semiEarly=False):
  
  if whichError=='minTrain':
    weights='tWeights'
  else:
    weights='vWeights'
  
  #repetitive caruana
  model=modelFunction()
  to_device(model, device)
  
  resultsArray=[]
  #1] selector  
  negativeLinspace=lambda xt,yt,xv,yv,error,model,optFunction: {'negLr':lrLinspace(xt,yt,xv,yv,error,model,optFunction,graph=False,negative=True,upperBound=2)['negative/duration'],'maxLr':lrLinspace(xt,yt,xv,yv,error,model,optFunction,graph=False)['minSteep']} #dejo un upperbound grande al principio porque es el primero de todos
  NEGparameterSelectorArray=[negativeLinspace for i in range(stochs)]
  #2] cycler
  newNegative=lambda epoch,semiEpoch,cycleParams: negativeLearnCaruana2(epoch,semiEpoch,cycleParams)
  NEGcyclerFunctionArray=[newNegative for i in range(stochs)]
  #3] epochs
  NEGsemiEpochsArray=[2**i for i in np.arange(7,min([stochs+7-2,14]),1)]+[2**7 for i in range(max([stochs-7,2]))]
  resamples=tunedCyclicFit(NEGparameterSelectorArray,NEGcyclerFunctionArray,NEGsemiEpochsArray, model, error, optFunction, xt,yt,xv,yv,graph=False,early=early,semiEarly=semiEarly)['results']
  
  #ahora chequeo elmodelo que llegue mas profundo
  for weighted in resamples:
    model.load_state_dict(weighted['vWeights']) #creo que no hay que hacerlo dinamico, creo quesiempre va minVal
    resultsArray.append(tunedCyclicFit(parameterSelectorArray,cyclerFunctionArray,semiEpochsArray, model, error, optFunction, xt,yt,xv,yv,graph=False,early=early,semiEarly=semiEarly))

  errores=[x[whichError] for x in resultsArray]
  minimoError=min(errores)
  indexMin=errores.index(minimoError)
  bestResult=resultsArray[indexMin]

  return {'array':errores,'minError':minimoError,'bestModel':bestResult[weights]}
```
